```python
data=df.iloc[:,0:11]
data
```

```python
#Using PCA alongs the columns
from sklearn.decomposition import PCA
from sklearn import preprocessing

pca = preprocessing.scale(data)

pca = PCA(n_components=11)
pc = pca.fit_transform(data)
names = ['pc1','pc2','pc3','pc4','pc5','pc6','pc7','pc8','pc9','pc10','pc11']
pcc = pd.DataFrame(data = pc, columns = names)
```

```python
pcc
```

```python
###Importing KMean Cluster
```

```python
from sklearn.cluster import KMeans 
```

```python
wcss=[]
for k in range(1,15):
    kmeans=KMeans(n_clusters=k,init='k-means++')
    kmeans.fit_transform(data)
    wcss.append(kmeans.inertia_)
    
    
    
```

```python
plt.plot(range(1,15),wcss)#finding elbow value
plt.xticks(range(1,15))
plt.xlabel("Number of cluster")
plt.ylabel("WCSS")
plt.show()
```

```python
Kmeans=KMeans(n_clusters=4,init="k-means++")
y_labels=Kmeans.fit_predict(data)
```

```python
#CLuster Visualization
df["cluster_num"]=kmeans.labels_
sns.scatterplot(data=pcc, x="pc1", y="pc2", hue=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], 
            marker="X", c="r", s=80, label="centroids")
plt.grid(True)
plt.legend()
plt.show()
```

```python
#Audience in each cluster based on gender
```

```python
sns.boxplot(x=kmeans.labels_, y="Age", data=df)#mean of Visit Frequency of each Cluster
```

```python
#Calculating the mean
#Visit frequency

df['VisitFrequency'] = LabelEncoder().fit_transform(df['VisitFrequency'])
visit_freq = df.groupby('cluster_num')['VisitFrequency'].mean()
visit_freq = visit_freq.to_frame().reset_index()
visit_freq
```

```python
#Range of visit frequency (2.4-2.8)
#Highest Visit frequency of auidience in cluster 12
```

```python
df['Like'] = df['Like'].astype('int')
```

```python
df['Like'] = LabelEncoder().fit_transform(df['Like'])
Like = df.groupby('cluster_num')['Like'].mean()
Like = Like.to_frame().reset_index()
Like
```

```python
df['Like'].value_counts()
```

```python
Like=Like.drop('cluster_num',axis=1)
```

```python
concate=pd.concat([visit_freq,Like],axis=1) # Joining two dataframes
concate
```

```python
plt.figure(figsize = (9,4))
sns.scatterplot(x = "VisitFrequency", y = "Like",data=concate,s=400, color="r")
plt.title("Simple segment evaluation plot for the fast food data set",
          fontsize = 15) 
plt.xlabel("Visit", fontsize = 12) 
plt.ylabel("Like", fontsize = 12) 
plt.grid(True)
plt.tight_layout()
plt.show()
```
